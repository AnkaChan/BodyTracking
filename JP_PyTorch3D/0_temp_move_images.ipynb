```python
import glob
import os
import shutil

```

# sub folders by camera name

```python
in_dir = r'Z:\2020_03_22_NewSuitDesignCapture\Converted'
out_dir = r'D:\1_Projects\200325_PyTorch3d_Toy\16b_data\input\images'
        
cams = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L', 'M', 'N', 'O', 'P']
for c in cams:
    d = out_dir + '\\{}'.format(c)
    if not os.path.exists(d):
        os.mkdir(d)
        print(d)

f0 = 3180
for i in range(0, 1):
    fn = f0 + i
    for c in cams:
        f_path = in_dir + r'\\{}\{}0{}.pgm'.format(c, c, fn)
        o_path = out_dir + r'\\{}\{}0{}.pgm'.format(c, c, fn)
        shutil.copyfile(f_path, o_path)
print('Done')
```

# sub folders by image name

```python
in_dir = r'Z:\2019_12_13_Lada_Capture\Converted'
out_dir = r'Z:\2019_12_13_Lada_Capture\200512_MeshSparseFitted\images'
        
cams = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L', 'M', 'N', 'O', 'P']

image_names = ['03067', '04735', '06250', '06550']
for img_name in image_names:
    out_dir_curr = out_dir + r'\{}'.format(img_name)
    if not os.path.exists(out_dir_curr):
        os.mkdir(out_dir_curr)

    for c in cams:
        f_path = in_dir + r'\\{}\{}{}.pgm'.format(c, c, img_name)
        o_path = out_dir_curr + r'\\{}{}.pgm'.format(c, img_name)
        shutil.copyfile(f_path, o_path)
        print(o_path)
print('Done')
```
